**1. Fetching Multiple Repositories at Once**

If you have many different Git repositories (e.g., repo1, repo2, repo3) inside a common parent folder, you can use a loop or the find command combined with the -C flag to run git fetch --all on all of them simultaneously.

```r
Get-ChildItem -Directory | ForEach-Object {
    if (Test-Path "$($_.FullName)\.git") {
        Write-Host "Pulling latest changes for $($_.Name)..."
        git -C $_.FullName pull
    }
}
```

This script:

* Gets all directories (-Directory) in the current parent folder.
* Checks if a .git folder exists inside that directory.
* If it does, it runs git fetch --all inside that repository using the -C option.

**2. Bring an Existing GitHub Repo Local (Clone)**

If you already created the repository on the GitHub website (or want to pull down a framework/tutorial repository), navigate to your main container directory and clone it:

```r
# 1. Make sure you are in your main project folder
cd /path/to/your/main/directory

# 2. Clone the repository using HTTPS or SSH
git clone https://github.com/YOUR-USERNAME/NEW-REPO-NAME.git

# 3. Step inside your new local folder
cd NEW-REPO-NAME
```

**3. Connecting a New Local Project to GitHub (Push Existing)**

```r
# 1. Step into your existing project folder
cd /path/to/your/local/project

# 2. Initialize a local Git repository
git init

# 3. Stage and commit all existing files
git add .
git commit -m "Initial commit: local project setup"

# 4. Set the default branch to 'main'
git branch -M main

# 5. Link your local repository to a freshly created, empty GitHub repo
git remote add origin https://github.com/YOUR-USERNAME/NEW-REPO-NAME.git

# 6. Push your local history up to GitHub and track the remote branch
git push -u origin main
```

**4. Branching & Workspace Isolation**

```r
# Create a new branch and switch to it immediately
git checkout -b feature/new-trimming-params

# Switch back to an existing branch (e.g., main)
git checkout main

# List all local branches (the one with an asterisk * is your active branch)
git branch

# Delete a branch locally once it has been merged
git branch -d feature/new-trimming-params
```

**5. Undoing Changes & Saving Your Work**

**5. a) Discarding Uncommitted Changes**

```r
# Revert a specific file back to its last committed state (discarding unstaged changes)
git checkout -- filter_script.R

# Reset a staged file back to unstaged, keeping your file modifications intact
git reset HEAD filter_script.R
```

**5.b) The "Safety Net": Git Stash**

```r
# Save your uncommitted changes to a temporary storage shelf
git stash

# Switch branches, do your work, switch back, and then restore your saved changes:
git stash pop

# List all stashed changes
git stash list
```

**6. Comparing Changes (Diffing)**

```r
# Compare your unstaged changes in the working directory against your last commit
git diff

# Compare a specific file's current state with your last commit
git diff workflow_objectives.md

# Compare changes that you have already staged (run after 'git add')
git diff --staged
```

**7. Viewing History & Collaboration**

```r
# View a clean, single-line history of all commits
git log --oneline --graph --decorate

# See who modified each line of a specific file and when (perfect for finding out when a bug was introduced)
git blame raw_qc_pipeline.sh
```

**8. Cleaning Up Untracked Files**

```r
# Dry run: See what untracked files will be permanently deleted
git clean -n

# Force delete all untracked files and directories (use with caution!)
git clean -fd
```
